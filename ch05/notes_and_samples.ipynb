### Type Constraints in Haskell: Keeping Things Tidy

In Haskell, type constraints are like rules that guarantee certain types behave as expected within a function or data structure. Think of them as contracts: the function promises it can work with specific types, and the compiler checks if you hold up your end by providing valid inputs. This helps catch errors early and ensures your code is robust and reliable.

Let's look at an example:

```haskell
-- Function expects two numeric arguments and returns their sum
sum :: Num a => a -> a -> a
sum x y = x + y
```

Here, `Num a =>` is the type constraint. It states that the function `sum` can work with any type `a` **as long as that type is an instance of the `Num` class**. The `Num` class provides operations like addition (`+`), subtraction (`-`), and multiplication (`*`) for numeric types like `Int`, `Float`, etc.

The constraint ensures that:

1. You can only call `sum` with arguments of types that support these operations. Trying to add strings using `sum` wouldn't make sense, and the constraint prevents such errors.
2. The compiler knows what operations are available on type `a`, allowing it to optimize the code accordingly.

Here are some other commonly used type constraints:

* `Eq a`: The type `a` supports equality checks (`==` and `/=`).
* `Ord a`: The type `a` can be ordered (`<`, `>`, etc.).
* `Show a`: The type `a` can be converted to a string for printing.

**Remember:**

* Constraints apply to specific parts of a function or data structure. In the example above, the constraint only applies to arguments and the return type of `sum`.
* You can have multiple constraints separated by commas. For example, `(Num a, Ord a) =>` requires `a` to be both numeric and orderable.
* Some advanced features like type families and GADTs allow even more complex constraints.

Using type constraints effectively makes your Haskell code more readable, maintainable, and less prone to errors. So, embrace the power of contracts and keep your types in check!

### Ploymorphism

#### Constrained Polymorphism

Constrained polymorphism, also known as ad-hoc polymorphism, allows functions to work with different types that share specific properties. They achieve this flexibility through **type classes**, which define a set of operations that types can implement.

**Example:**

Consider a `Comparable` class with a single method `compare`:

```haskell
class Eq a where
  (==) :: a -> a -> Bool
  (<) :: a -> a -> Bool
```

Any type that implements `Eq` must provide implementations for both `==` and `<`. This allows functions like `sort` to work with different comparable types like `Int`, `String`, or custom data structures:

```haskell
sort :: Ord a => [a] -> [a]
sort xs = quicksort xs

quicksort :: Ord a => [a] -> [a]
quicksort [] = []
quicksort (pivot:xs) =
  let smaller = filter (< pivot) xs
      larger = filter (> pivot) xs
  in quicksort smaller ++ [pivot] ++ quicksort larger
```

`sort` uses the `Ord` type class, a subclass of `Eq` that adds the `<` operator. This constrains the function to only work with types that can be compared for ordering.

**Benefits:**

* **Code reuse:** Write generic functions that work with various types that share the same properties.
* **Type safety:** Type classes ensure used types satisfy required operations.
* **Improved extensibility:** Easily add new types by implementing required methods.

#### Parametric Polymorphism

Parametric polymorphism, also known as generics, allows functions to work with any type by using type variables. These variables act as placeholders, and the function can be applied to any concrete type.

**Example:**

Define a `map` function that applies a function to each element of a list:

```haskell
map :: (a -> b) -> [a] -> [b]
map f xs = [f x | x <- xs]
```

`map` takes two arguments:

* `f`: A function that takes an element of type `a` and returns an element of type `b`.
* `xs`: A list of elements of type `a`.

The function iterates through the list, applying `f` to each element and building a new list of the transformed elements. This works with any type `a` and `b` as long as the function `f` can handle them.

**Benefits:**

* **General-purpose code:** Write functions that work with any type without modification.
* **Type abstraction:** Focus on logic without specifying concrete types.
* **Reduced code duplication:** Avoid writing similar functions for different types.

#### Key Differences

| Feature           | Constrained Polymorphism | Parametric Polymorphism |
| ----------------- | ------------------------ | ------------------------ |
| Type restriction   | Requires types to implement specific operations | Works with any type |
| Flexibility       | Less flexible, limited to types in a class | More flexible, usable with any type |
| Expressiveness    | Captures relationships between types and operations | Focuses on generic behavior |
| Example           | `sort` works with all `Ord` types | `map` works with any list type |

Choosing between these depends on your specific needs. Constrained polymorphism is useful when types share specific functionality, while parametric polymorphism offers broader applicability at the expense of requiring type-agnostic logic.
